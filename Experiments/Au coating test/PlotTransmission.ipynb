```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Total coating time (min)	Transmission measured at 980nm (%)	Transmission measured at 785nm (%)	Transmission measured at 532nm (%)
# 0,00	100,00	100,00	100,00
# 3,67	49,82	49,46	49,52
# 7,62	29,59	28,62	30,05
# 12,00	11,46	11,88	11,43
# 16,20	4,09	4,02	4,00
# 19,27	1,75	1,76	1,76

CoatingTime = [0.00, 3.67, 7.62, 12, 16.2, 19.27]
Transmission = {
    980: [100, 49.82, 29.59, 11.46, 4.09, 1.75],
    785: [100, 49.46, 28.62, 11.88, 4.02, 1.76],
    532: [100, 49.52, 30.05, 11.43, 4.00, 1.76]
}

plt.figure()
for wl in Transmission.keys():
    plt.semilogy(CoatingTime,Transmission[wl],label=f"at {wl} nm",marker='x',linewidth=1)
plt.xlabel("Coating time (min)")
plt.ylabel("Transmission (%)")
plt.title("Transmission (%) measured by the photodiode sensor \n as a function of Total coating time (min)")
plt.legend()
plt.ylim(bottom=1)
# plt.show()
plt.savefig("GoldCoatingTest.pdf",bbox_inches="tight")
```
